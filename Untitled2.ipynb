```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('data.csv')

# Visualize Gender Distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()

# Visualize Distribution of Playing Sport
sport_play_counts = df['Do you play a sport?'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(sport_play_counts, labels=sport_play_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Do You Play a Sport?')
plt.show()

# Analyze GPA by Sport Participation
# First, ensure your GPA column is numeric
df['What is your C-GPA?'] = pd.to_numeric(df['What is your C-GPA?'], errors='coerce')

# Then, calculate and display mean GPA for those who play sports vs. those who don't
mean_gpa_sport = df.groupby('Do you play a sport?')['What is your C-GPA?'].mean()
print(f"Mean GPA by Sport Participation:\n{mean_gpa_sport}\n")

# For a more detailed analysis, you could compare GPA distributions between groups
# or conduct hypothesis testing to see if differences are statistically significant.

# If interested in the impact of playing sports on GPA as perceived by students
impact_counts = df['Does playing a sport affect your GPA in a bad way?'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(impact_counts, labels=impact_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Perceived Impact of Playing Sports on GPA')
plt.show()

# Optional: Analyze "What sport do you play?" if you're interested in the distribution of sports
# Be mindful of how many unique sports are listed to decide between a pie or a bar chart.
# Encoding "Do you play a sport?" into numeric format for correlation analysis
df['Play_Sport_Numeric'] = df['Do you play a sport?'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculating Pearson correlation coefficient
correlation = df[['Play_Sport_Numeric', 'What is your C-GPA?']].corr(method='pearson')
print(f"Pearson Correlation Coefficient between playing a sport and C-GPA:\n{correlation}")
# Standard Deviation of GPA
std_gpa = df['What is your C-GPA?'].std()
print(f"Standard Deviation of C-GPA: {std_gpa}")

```
